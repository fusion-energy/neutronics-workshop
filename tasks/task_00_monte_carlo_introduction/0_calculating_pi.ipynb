# Part 1 - Calculating Pi using Monte Carlo

Monte Carlo used the production of a statistically significant number of randomly produced events to calculate a feature of the system. 

A simple example of Monte Carlo is using random numbers to calculate the mathematical constant pi. 

An array can be created, containing $n$ randomly distributed data points in a square of diameter $2$ between $-1$ and $1$:

```python
from utils import create_xy_array, calculate_pi
xy = create_xy_array(5000)
print(xy[:2,:])
```

All of the randomly created points will reside within the area of this square, however a portion of them will be within an inner circular area, centered on ($0$,$0$), also with a diameter, $d$, of $2$. 

The cartesian points (x,y) which reside within the circle can be calculated used the equation of a circle: 

$$ x^2+y^2 < \Bigg(\frac{d}{2}\Bigg)^2 $$

$$ x^2+y^2 < 1 $$

```python
from utils import plot_circle
plot_circle(xy)
```

As 100% of the data points lie within the area of the square, the fraction of points which lie within the area of the circle is equal to the ratio of the area of the circle over the area of the square. 

(This statement is true for very large number of data points, reducing in accuracy with lower statistics and breaking down entirely with very few statistics).

$$ A_{square} = d^2$$

$$ A_{circle} = \frac{1}{4}\pi d^2 $$

$$ \frac{A_{circle}}{A_{square}} = \frac{\frac{1}{4}\pi d^2}{d^2} = \frac{\pi}{4} $$

```python
xy = create_xy_array(10000000)
calculate_pi(xy)
```
