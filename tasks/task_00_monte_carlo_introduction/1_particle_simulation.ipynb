# Part 2 - Particle Simulation

When simulating the transport of a neutron through a material, there are many reactions which will occur throughout the lifetime of the particle. Even if only accounting for scattering events of a monoenergetic particle, there are a large number of variables: The scattering angle at each interaction and the distance between scattering events. 

A simple model of a neutron travelling through matter is shown below, where the mean and spread on the scattering angle are variables, as well as the distance between scattering events and the variance on this: 

```python
from utils import plot_track, create_particle_track

particle_life_time = 1000
mean_angle = 30
angle_spread = 14 
mean_free_path = 5 
mean_free_path_spread = 1
number_of_tracks = 1
distribution_type = 'gauss' #'uniform' or 'gauss'

track_list = []
for i in range(number_of_tracks): 
    track_list.append(create_particle_track(particle_life_time, mean_angle, angle_spread, mean_free_path, mean_free_path_spread, distribution_type))
plot_track(track_list)
```
